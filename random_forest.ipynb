```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

from ipynb.fs.defs.Naive_bayes import categorical_enconding
```

```python
credit = pd.read_csv("../Credit.csv") ; credit.head(3)
```

```python
#database split

class_ = credit['class']
attb = credit.drop('class', axis=1)

#categorical enconding
attb = categorical_enconding(attb)

#train test split
X_train, X_test, Y_train, Y_test = train_test_split(attb, class_,random_state=0, train_size = 0.7)

#model selection
random_forest = RandomForestClassifier(n_estimators=100) #defines the number of decision trees that will be constructed in the model

#model training 
random_forest.fit(X_train, Y_train)

#test
Y_predict = random_forest.predict(X_test)

#accuracy check
acc = accuracy_score(Y_test,Y_predict)
acc
```

```python
confusion_mat = confusion_matrix(Y_test,Y_predict)
confusion_mat
```
